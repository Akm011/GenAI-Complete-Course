List Comprehension
Basics Syantax [expression for item in iterable]

with conditional logic [expression for item in iterable if condition]

Nested List Comprehension [expression for item1 in iterable1 for item2 in iterable2]

```python
### Basic List Comphrension

sqaure=[num**2 for num in range(10)]
print(sqaure)
```

```python
### List Comprehension with Condition
even_numbers=[num for num in range(10) if num%2==0]
print(even_numbers)
```

```python
## Nested List Comphrension

lst1=[1,2,3,4]
lst2=['a','b','c','d']

pair=[[i,j] for i in lst1 for j in lst2]

print(pair)
```

```python
## List Comprehension with function calls
words = ["hello", "world", "python", "list", "comprehension"]
lengths = [len(word) for word in words]
print(lengths)  # Output: [5, 5, 6, 4, 13]
```

Sets
Sets are a built-in data type in Python used to store collections of unique items. They are unordered, meaning that the elements do not follow a specific order, and they do not allow duplicate elements. Sets are useful for membership tests, eliminating duplicate entries, and performing mathematical set operations like union, intersection, difference, and symmetric difference.

```python
my_empty_set=set([1,2,3,6,5,4,5,6])
print(my_empty_set)
```

```python
my_set=set([1,2,3,4,5,6])
print(my_set)
## Basics Sets Operation
## Adiing and Removing Elements
my_set.add(7)
print(my_set)

## Remove the elements from a set
my_set.remove(3)
print(my_set)

## pop method
removed_element=my_set.pop()
print(removed_element)
print(my_set)

## Set Memebership test
my_set={1,2,3,4,5}
print(3 in my_set)
print(10 in my_set)
```

```python
## MAthematical Operation
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

### Union
union_set=set1.union(set2)
print(union_set)

## Intersection
intersection_set=set1.intersection(set2)
print(intersection_set)

set1.intersection_update(set2)
print(set1)
```

Introduction to Dictionaries
Dictionaries are unordered collections of items. They store data in key-value pairs. Keys must be unique and immutable (e.g., strings, numbers, or tuples), while values can be of any type.

```python
student={"name":"Krish","age":32,"grade":24}
print(student)
print(type(student))
```

```python
# Single key is slways used
student={"name":"Krish","age":32,"name":24}
print(student)
```

```python
## Dictionary methods

keys=student.keys() ##get all the keys
print(keys)
values=student.values() ##get all values
print(values)

items=student.items() ##get all key value pairs
print(items)
```

```python
## Dictionary Comphrehension
squares={x:x**2 for x in range(5)}
print(squares)
```

```python
## Merge 2 dictionaries into one

dict1={"a":1,"b":2}
dict2={"b":3,"c":4}
merged_dict={**dict1,**dict2}
print(merged_dict)
```

```python
## shallow copy 1

#You're not copying the dictionary here — you're just creating a new reference (alias) to the same object in memory
#student_copy and student both point to the same dictionary object.
student={"name":"Krish","age":32,"grade":'A'}
student_copy=student
print(student)
print(student_copy)

student["name"]="Krish2"
print(student)
print(student_copy)

print("Checking the ID's")
print(id(student))
print(id(student_copy)) 
```

```python
## shallow copy 2
# student.copy() creates a shallow copy of the dictionary — a new dictionary with the same key-value pairs,
# but since the values are all immutable (str, int, char), shallow vs deep doesn’t really matter here.
student={"name":"Krish","age":32,"grade":'A'}
student_copy1=student.copy() 
print(student_copy1)
print(student)

student["name"]="KRish3"
print(student_copy1)
print(student)

print("Checking the ID's")
print(id(student))
print(id(student_copy)) 
```

dict.copy() makes a shallow copy.

For dictionaries with immutable values, shallow copy behaves just like a deep copy.

Only changes to mutable values (like lists or other dicts) would reflect in both if modified.

Examples of immutable types:

int
float
str
tuple
bool

Examples of mutable types:

list
dict
set

```python
import copy

# Original dictionary with a nested mutable list
student = {
    "name": "Krish",
    "age": 32,
    "grades": [90, 85, 88]  # mutable list
}
print("Original before change:", student)
# Perform a deep copy
student_deepcopy = copy.deepcopy(student)
student_shallowcopy = student.copy()

# Change the original
student["grades"][0] = 100

# Print both dictionaries
print("Original after change:", student)
print("Deep Copy:", student_deepcopy)
print("Shallow Copy:", student_shallowcopy)
```
