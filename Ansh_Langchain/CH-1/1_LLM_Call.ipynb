```python
from  langchain_openai import ChatOpenAI
import os 
from dotenv import load_dotenv

# This function will load all the variables from the .env file and will 
# make them available in the os.environ dictionary (env variables)
load_dotenv() 

#os.environ  --> returns a dictionary of all the environment variables

if os.environ.get("OPENAI_API_KEY"):
    print("Bro API KEY Variable exists")
else:
    raise ValueError("OPENAI_API_KEY not found")

```

```python
llM_openapi = ChatOpenAI(model_name="gpt-5", temperature=0)
llM_openapi.invoke("Bro, tell me a fun fact").content
```

```python
from langchain.chat_models import init_chat_model

llm_openai = init_chat_model(model = "gpt-5-mini")

llm_openai.invoke("Hello, how are you?")
```
