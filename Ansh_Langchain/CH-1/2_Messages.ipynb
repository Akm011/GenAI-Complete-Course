# **MESSAGES**

```python
from  langchain_openai import ChatOpenAI
import os 
from dotenv import load_dotenv

# This function will load all the variables from the .env file and will 
# make them available in the os.environ dictionary (env variables)
load_dotenv() 

if os.environ.get("OPENAI_API_KEY"):
    print("Bro API KEY Variable exists")
else:
    raise ValueError("OPENAI_API_KEY not found")
```

```python
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
from  langchain_openai import ChatOpenAI

llm_openai = ChatOpenAI(model="gpt-5-mini",temperature=0)
my_messages = [
    SystemMessage(content="You are a gen-z assistant who always answers in a fun way"),
    HumanMessage(content="Bro, tell me a fun fact")
]

llm_openai.invoke(my_messages).content
```

# **Prompts**

Prompts are the messages that are sent to the LLM. Prompts are more user-friendly than messages.

```python
from langchain_core.prompts import PromptTemplate

user_input = input("Enter the topic of fun fact:  ")
dynamic_prompt = PromptTemplate.from_template("Write a fun fact about {topic}")
ready_prompt = dynamic_prompt.invoke({"topic": user_input})
# print(ready_prompt)
llm_openai.invoke(ready_prompt)
```

```python
# Below is same as [4] using ChatPromptTemplate.from_messages

from langchain_core.prompts import ChatPromptTemplate 

prompt_template = ChatPromptTemplate.from_messages([
    ("system","You are a {tone} assistant"),
    ("user","Write a fun fact about {topic}.")
])

user_input = input("Enter a topic: ")
user_tone = input("Enter a tone: ")

ready_prompt = prompt_template.invoke({"topic":user_input,"tone":user_tone})

llm_openai.invoke(ready_prompt.messages)
```
